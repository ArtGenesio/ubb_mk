```python
import numpy as np
import matplotlib.pyplot as plt

def elu(x, alpha=1.0):
    return np.where(x > 0, x, alpha * (np.exp(x) - 1))

def elu_gradient(x, alpha=1.0):
    return np.where(x > 0, 1.0, alpha * np.exp(x))

x = np.linspace(-5, 5, 200)
y = elu(x)
grad = elu_gradient(x)

plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Funkcja ELU ($f(x)$)', color='blue', linewidth=2)
plt.plot(x, grad, label='Gradient ELU ($f\'(x)$)', color='red', linestyle='--', linewidth=2)

plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.title('Funkcja aktywacji ELU i jej gradient', fontsize=14)
plt.xlabel('x', fontsize=12)
plt.ylabel('y', fontsize=12)
plt.grid(True, linestyle=':', alpha=0.6)
plt.legend(fontsize=12)
plt.show()
```
